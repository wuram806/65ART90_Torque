```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import sklearn.preprocessing
import os
import matplotlib.pyplot
```

```python
os.getcwd()
```

```python
torque_path=[]
for path,dir,files in os.walk("D:\\PythonBasic-main\\PythonBasic-main\\2022_Torque_Low"):
    for filename in files: #files 는 각 폴더의 파일명을 list로 반환함. filename in files 하면.. list 안의 각 원소 반환하기때문에
                            #파일이름만 보이게 됨!!!
        ext=os.path.splitext(filename)[1] #os.path.splitext(--) : --의 파일이름을 '.' (dot) 기준으로 나누어 튜플로 반환함.
        if ext == '.csv':
            torque_path_str=path+'\\'+filename
            torque_path.append(torque_path_str)
```

```python
pd.read_csv(torque_path[0],skiprows=6,usecols=[1],encoding='cp949')
```

```python
df=pd.DataFrame()
for i in range(0,len(torque_path)):
    df2=pd.read_csv(torque_path[i],skiprows=6,usecols=[1],encoding='cp949')
    df=pd.concat([df,df2],axis=1)
```

```python
df=df.transpose()
```

```python
df=df.reset_index()
```

```python
df=df.drop(columns='index')
```
